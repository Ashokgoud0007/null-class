```python
import numpy as np
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M',''))
    elif 'k' in size:
        return float(size.replace('k',''))/1024
    else:
        return np.nan
data['Size']=data['Size'].apply(convert_size)
```

```python
data.head()
```

```python
#Convert the Installs columns to numeric by removing commas and +
data['Installs']=data['Installs'].str.replace(',','').str.replace('+','').astype(int)

#Convert Price column to numeric after removing $
data['Price']=data['Price'].str.replace('$','').astype(float)
```

```python
data.head()
```

```python
# Filter the dataset based on the specified conditions
filtered_df = data[
    (data['Rating'] > 3.5) & 
    (data['Category'] == 'games') & 
    (data['Installs'] > 50000)
]
```

```python
# Filter the dataset based on the specified conditions
filtered_df = data[
    (data['Rating'] > 3.5) & 
    (data['Category'] == 'MEDICAL') & 
    (data['Installs'] > 50000)
]
```

```python
# Check the current time

# import datetime libruary
import datetime
current_time = datetime.datetime.now().time()
```

```python
current_time
```

```python
# Check if the current time is between 12 PM and 4 PM
if datetime.time(9, 0) <= current_time <= datetime.time(16, 0):
    # Create a bubble chart
    plt.figure(figsize=(12, 8))
    plt.scatter(
        filtered_df['Size'], 
        filtered_df['Rating'], 
        s=filtered_df['Installs'] / 1000,  # Bubble size (scaled down for better visualization)
        alpha=0.5,
        edgecolors="r",
        linewidth=0.5
    )

    plt.title('Bubble Chart of App Size vs Average Rating')
    plt.xlabel('App Size (MB)')
    plt.ylabel('Average Rating')
    plt.grid(True)
   # plt.xlim(0, filtered_df['Size'].max() + 10)  # Adjust x-axis limit
    #plt.ylim(o, 5)  # Rating ranges from 0 to 5
    plt.show()
else:
    print("The graph can only be displayed between 12 PM and 4 PM.")
```
