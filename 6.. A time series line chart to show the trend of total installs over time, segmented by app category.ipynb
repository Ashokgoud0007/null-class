##  A time series line chart to show the trend of total installs over time, segmented by app category

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
# droping duplicates 
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

```python
data.head()
```

```python
import numpy as np
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M',''))
    elif 'k' in size:
        return float(size.replace('k',''))/1024
    else:
        return np.nan
data['Size']=data['Size'].apply(convert_size)
```

```python
data.head()
```

```python
#Convert the Installs columns to numeric by removing commas and +
data['Installs']=data['Installs'].str.replace(',','').str.replace('+','').astype(int)

#Convert Price column to numeric after removing $
data['Price']=data['Price'].str.replace('$','').astype(float)
```

```python
data.head()
```

```python
# Convert 'date' to datetime
data['Last Updated'] = pd.to_datetime(data['Last Updated'])
```

```python
data.head()
```

```python
# Filter data for apps starting with 'E', content rating 'Teen', and installs > 10k
filtered_data = data[(data['App'].str.startswith('E')) & 
                     (data['Content Rating'] == 'Teen') & 
                     (data['Installs'] > 10000)]
```

```python
filtered_data
```

```python
#filtered_data.shape
filtered_data.shape
```

```python
# Create a month-year column for grouping
filtered_data['month'] = filtered_data['Last Updated'].dt.to_period('M')
```

```python
filtered_data
```

```python

# Group by month and category to sum installs
monthly_installs = filtered_data.groupby(['month', 'Category'])['Installs'].sum().reset_index()
```

```python
monthly_installs
```

```python
# Calculate month-over-month growth
monthly_installs['growth'] = monthly_installs.groupby('Category')['Installs'].pct_change()
```

```python
monthly_installs
```

```python
# Check the current time

# import datetime libruary
import datetime
current_time = datetime.datetime.now().time()
current_time
```

```python
# Check if the current time is between 12 PM and 4 PM
if datetime.time(16, 0) <= current_time <= datetime.time(20, 0):
    # Create a bubble chart
    # Plotting
    plt.figure(figsize=(14, 7))

    # Loop through each category to plot
    for category in monthly_installs['Category'].unique():
        category_data = monthly_installs[monthly_installs['Category'] == category]
        plt.plot(category_data['month'].astype(str), category_data['Installs'], label=category)
    
        # Highlight significant growth (more than 20%)
        significant_growth = category_data[category_data['growth'] > 0.2]
        if not significant_growth.empty:
            plt.fill_between(significant_growth['month'].astype(str), 
                             significant_growth['Installs'], 
                             alpha=0.3)

    # Chart details
    plt.title('Total Installs Over Time by App Category')
    plt.xlabel('Month')
    plt.ylabel('Total Installs')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.grid()
    plt.show()
else:
    print("The graph can only be displayed between 12 PM and 4 PM.")
```
