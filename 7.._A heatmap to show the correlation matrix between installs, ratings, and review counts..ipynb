```python
data.dropna(inplace=True)
```

```python
# droping duplicates 
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

```python
data.head()
```

```python
import numpy as np
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M',''))
    elif 'k' in size:
        return float(size.replace('k',''))/1024
    else:
        return np.nan
data['Size']=data['Size'].apply(convert_size)
```

```python
data.head()
```

```python
#Convert the Installs columns to numeric by removing commas and +
data['Installs']=data['Installs'].str.replace(',','').str.replace('+','').astype(int)

#Convert Price column to numeric after removing $
data['Price']=data['Price'].str.replace('$','').astype(float)
```

```python
data.head()
```

```python
# importing the necesary libruary
from datetime import datetime, timedelta
```

```python

# Convert 'last_updated' to datetime
data['Last Updated'] = pd.to_datetime(data['Last Updated'])
```

```python
data.head()
```

```python
data["Reviews"].values
```

```python
# reviews has objecct dtype and now converts into int dtype
data['Reviews'] = pd.to_numeric(data['Reviews'], errors='coerce')
```

```python
data["Reviews"].dtypes
```

```python
# Filter the data

# note
# in this one year before data is not appear. so,i took 7 years before data 
seven_year_ago = datetime.now() - timedelta(days=2555)
filtered_df = data[
    (data['Last Updated'] >= seven_year_ago) &
    (data['Installs'] >= 100000) &
    (data['Reviews'] >= 1000) &
 #   (data['Rating'])&
    (~data['Genres'].str.startswith(('A', 'F', 'E', 'G', 'I', 'K')))
]
```

```python
seven_year_ago
```

```python
filtered_df
```

```python
# Calculate the correlation matrix
#filtered_df['Installs'] = filtered_df['Installs'].str.replace(',', '').astype(int)  # Ensure installs are integers
correlation_matrix = filtered_df[['Installs', 'Rating', 'Reviews']].corr()
correlation_matrix
```

```python
# Check the current time
current_time = datetime.now().time()
if current_time >= datetime.strptime('15:00', '%H:%M').time() and current_time <= datetime.strptime('18:00', '%H:%M').time():
    # Generate the heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap='viridis', vmin=-1, vmax=1)
    plt.title('Correlation Heatmap between Installs, Ratings, and Review Counts')
    plt.show()
else:
    print("The graph can only be generated between 3 PM and 6 PM.")
```
