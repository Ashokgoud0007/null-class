## Use a grouped bar chart to compare the average rating and total review count for the top 10 app categories by number of installs

```python
# import necesary libruaries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
data["Rating"]=data["Rating"].fillna(data["Rating"].median())
```

```python
#ploting, to check the rating category to fill the null values 
import seaborn as sns
sns.displot(data.Rating)
```

```python
# we shoild apply drop method for remaining categories because it has low null values
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
# droping duplicates 
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

```python
import numpy as np
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M',''))
    elif 'k' in size:
        return float(size.replace('k',''))/1024
    else:
        return np.nan
data['Size']=data['Size'].apply(convert_size)
```

```python
data.head()
```

```python
#Convert the Installs columns to numeric by removing commas and +
data['Installs']=data['Installs'].str.replace(',','').str.replace('+','').astype(int)

#Convert Price column to numeric after removing $
data['Price']=data['Price'].str.replace('$','').astype(float)
```

```python
data.head()
```

```python
#Revenue column
data['Revenue']=data['Price']*data['Installs']
```

```python
data.head()
```

```python
data['Last Updated'] = pd.to_datetime(data['Last Updated'])
```

```python
data.head()
```

```python
# Filter the data
filtered_data= data[
    (data['Rating'] >= 4.0) &
    (data['Size'] >= 10) &
    (data['Last Updated'].dt.month == 1)
]
```

```python
filtered_data
```

```python
# Group by category and calculate average rating and total reviews
grouped_data = filtered_data.groupby('Category').agg(
    Average_Rating=('Rating', 'mean'),
    Total_Reviews=('Reviews', 'sum'),
    Total_Installs=('Installs', 'sum')
).reset_index()
```

```python
grouped_data
```

```python
# Getting the top 10 categories by installs
top_categories = grouped_data.nlargest(10, 'Total_Installs')
```

```python
top_categories
```

```python
# Check current time
 # Ensure this import is included
from datetime import datetime 


# Check current time

current_time = datetime.now().time()
if current_time >= datetime.strptime("10:00", "%H:%M").time() and current_time <= datetime.strptime("17:00", "%H:%M").time():
    # Create a grouped bar chart
    plt.figure(figsize=(12, 6))
    bar_width = 0.35
    index = range(len(top_categories))  # Define index here

    # Plotting
    plt.bar(index, top_categories['Average_Rating'], bar_width, label='Average Rating', color='b')
    plt.bar([i + bar_width for i in index], top_categories['Total_Reviews'], bar_width, label='Total Reviews', color='g')

    # Adding labels and title
    plt.xlabel('App Categories')
    plt.ylabel('Values')
    plt.title('Average Rating and Total Review Count for Top 10 App Categories')
    plt.xticks([i + bar_width / 2 for i in index], top_categories['Category'], rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()
else:
    print("The graph can only be generated between 10 AM and 5 PM.")
```
