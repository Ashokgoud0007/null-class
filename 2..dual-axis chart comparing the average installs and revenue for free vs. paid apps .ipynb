```python
import numpy as np
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M',''))
    elif 'k' in size:
        return float(size.replace('k',''))/1024
    else:
        return np.nan
data['Size']=data['Size'].apply(convert_size)
```

```python
# Clean function with regex


import re


def clean_Android_Ver(Android_Ver):
    return re.sub(r'[^0-9.]', '', Android_Ver)  # Keep only numbers and dots

# Apply to the column
data1['clean_Android_Ver'] = data1['Android_Ver'].apply(clean_Android_Ver)
data1.head()
```

```python
# converting the version values into numeric
```

```python
def version_to_numeric(clean_Android_Ver):
    # Remove non-numeric characters except for dots
    clean_Android_Ver = ''.join(c if c.isdigit() or c == '.' else '' for c in clean_Android_Ver)
    
    # Split the version string
    parts = clean_Android_Ver.split('.')
    
    # Normalize length to ensure we have at least 3 components
    while len(parts) < 3:
        parts.append('0')
    
    # Convert to integers, ensuring we handle any empty strings
    major = int(parts[0]) if len(parts) > 0 and parts[0] else 0
    minor = int(parts[1]) if len(parts) > 1 and parts[1] else 0
    patch = int(parts[2]) if len(parts) > 2 and parts[2] else 0
    
    # Calculate numeric value
    return major * 10000 + minor * 100 + patch
data1["numeric_value"]=data1["clean_Android_Ver"].apply(version_to_numeric)
```

```python
# converting the nnumeric_value into with decimal point
data1["numeric_value"]=data1["numeric_value"]/10000
```

```python
# Apply filters
filtered_data1 = data1[
    (data1['Installs'] > 10000) &
    (data1['Revenue'] > 10000) &
    (data1['numeric_value'] > 4.0000) &
    (data1['Size'] > 15) &
    (data1['Content Rating'] == 'Everyone') &
    (data1['App'].str.len() <= 30)
]
```

```python

# Identify top 3 categories by the number of apps
top_categories = filtered_data1['Category'].value_counts().nlargest(3).index
```

```python
# Filter for only the top categories
filtered_data1 = filtered_data1[filtered_data1['Category'].isin(top_categories)]
```
